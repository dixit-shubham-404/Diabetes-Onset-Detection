```python
#import neccesary libraries for handling dataset
import pandas as pd
import numpy as np
```

```python
#dataset extracted from kaggle
df=pd.read_csv('diabetes-dataset.csv')

#analyzing database
df.describe()
```

By analyzing the result we can see there are lots for parameters whose value is zero which is not possible

```python
#Now by analyzing all this we replace the zero with NaN to get better reseult
columns=['Glucose','BloodPressure','SkinThickness','Insulin','BMI']

for col in columns:
    df[col].replace(0,np.NaN,inplace=True)

df.describe()
```

```python
#Now droping all those rows with NaN value as it can affect our model
df.dropna(inplace=True)
df.describe()
#Now e have total of 392 valid patient data
```

```python
#converting datafram to numpy array so to perform the opertion easily
dataset=df.values
print(dataset)
```

```python
print(dataset.shape)
```

```python
#splitting the dataset into x(input) and y(output)
x=dataset[:,0:8]
y=dataset[:,8].astype(int) #changing float to int as it represent class only
print(x.shape)
print(y.shape)
#we have total 8 input parameter and 1 output parameter(0=non-diabetic 1-diabetic)
```

```python
#if we see the dataset the range of value for different attribute are different
#Some are big and some are small
#To avoid the model giving the importance to only big attribtes
#we normalize the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_standard=scaler.fit_transform(x)
data = pd.DataFrame(x_standard)
data.describe()
```

```python
#importing necceasary libraries for creating model
from sklearn.model_selection import GridSearchCV,KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import Adam
```

```python
#In this section we will optimize the batch size and epochs for the model

seed=6
np.random.seed(seed)

#create model-In this we would be using sequential model with three dense layers 
def create_model():
    model=Sequential()
    model.add(Dense(8,input_dim=8,kernel_initializer='normal',activation='relu'))
    model.add(Dense(4,input_dim=8,kernel_initializer='normal',activation='relu'))
    model.add(Dense(1,activation='sigmoid')) #activation sigmoid is used for binary classification
    
    adam=Adam(lr=0.01) #learning rate
    model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
    return model

model=KerasClassifier(build_fn=create_model,verbose=0)


#defining grid search parameters
batch_size=[10,20,40]
epochs=[10,50,100]

#make a dictionary of the grid search parameters
param_grid=dict(batch_size=batch_size,epochs=epochs)

#build and fit grid search
grid=GridSearchCV(estimator=model, param_grid=param_grid, cv= KFold(random_state=seed), verbose=10)
grid_result = grid.fit(x_standard,y)

#analyzing the results
print('Best: {0}, using {1}'.format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_['mean_test_score']
stds=grid_result.cv_results_['std_test_score']
params=grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0}  ({1}) with: {2}'.format(mean,stdev,param))

#By nalyzing the result the best batch_size is 10 and number of epoch is 50
```

```python
#In this section we will optimize the dropout rate and learning rate
from tensorflow.keras.layers import Dropout

seed=6
np.random.seed(seed)


def create_model(learn_rate,dropout_rate):
    model=Sequential()
    model.add(Dense(8,input_dim=8,kernel_initializer='normal',activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(4,input_dim=8,kernel_initializer='normal',activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1,activation='sigmoid')) #activation sigmoid is used for binary classification
    
    adam=Adam(lr=learn_rate) #learning rate
    model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
    return model


#create model
model=KerasClassifier(build_fn=create_model,epochs=50,batch_size=10,verbose=0)

#param_grid parameters
learn_rate=[0.001,0.01,0.1]#shows how fast or slow we are updating the parameters
dropout_rate=[0.0,0.1,0.2]

#make a dictionary of the grid search parameters
param_grid=dict(learn_rate=learn_rate,dropout_rate=dropout_rate)

#build and fit grid search
grid=GridSearchCV(estimator=model, param_grid=param_grid, cv= KFold(random_state=seed), verbose=10)
grid_result = grid.fit(x_standard,y)

#analyzing the results
print('Best: {0}, using {1}'.format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_['mean_test_score']
stds=grid_result.cv_results_['std_test_score']
params=grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0}  ({1}) with: {2}'.format(mean,stdev,param))

#The results show we don't need droput layer and the best lerning rate is 0.01
```

```python
#By using the optimized parmeters we get above 
#In this section we will optimize the activation and kernel_initializer

seed=6
np.random.seed(seed)

#creating model and passing parameters we want to optimize
def create_model(activation,kernel_init):
    model=Sequential()
    model.add(Dense(8,input_dim=8,kernel_initializer=kernel_init,activation=activation))
    model.add(Dense(4,input_dim=8,kernel_initializer=kernel_init,activation=activation))
    model.add(Dense(1,activation='sigmoid')) #activation sigmoid is used for binary classification
    
    adam=Adam(lr=0.01) #learning rate
    model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
    return model


#initializing model
model=KerasClassifier(build_fn=create_model,epochs=100,batch_size=20,verbose=0)

#defining param_grid parameters
activation=['relu','linear','tanh','softmax']
kernel_init=['uniform','normal','zero']

#make a dictionary of the grid search parameters
param_grid=dict(activation=activation,kernel_init=kernel_init)

#build and fit grid search
grid=GridSearchCV(estimator=model, param_grid=param_grid, cv= KFold(random_state=seed), verbose=10)
grid_result = grid.fit(x_standard,y)

#analyzing the results
print('Best: {0}, using {1}'.format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_['mean_test_score']
stds=grid_result.cv_results_['std_test_score']
params=grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0}  ({1}) with: {2}'.format(mean,stdev,param))

#The result shows the best activation is 'linear' and kernel_initializer is 'normal'
```

```python
#By using the optimized parameter we get above 
#In this section we will optimize the number of neurons

seed=6
np.random.seed(seed)

#Creating model by passing number of neurons for both layer
def create_model(neuron1,neuron2):
    model=Sequential()
    model.add(Dense(neuron1,input_dim=8,kernel_initializer='normal',activation='linear'))
    model.add(Dense(neuron2,input_dim=neuron1,kernel_initializer='normal',activation='linear'))#its input will be output from previous layer
    model.add(Dense(1,activation='sigmoid')) #activation sigmoid is used for binary classification
    
    adam=Adam(lr=0.001) #learning rate
    model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
    return model


#create model
model=KerasClassifier(build_fn=create_model,epochs=100,batch_size=20,verbose=0)

#defining param_grid parameters
neuron1=[4,8,16]
neuron2=[2,4,8]

#make a dictionary of the grid search parameters
param_grid=dict(neuron1=neuron1,neuron2=neuron2)

#build and fit grid search
grid=GridSearchCV(estimator=model, param_grid=param_grid, cv= KFold(random_state=seed), refit=True,verbose=10)
grid_result = grid.fit(x_standard,y)

#analyzing the results
print('Best: {0}, using {1}'.format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_['mean_test_score']
stds=grid_result.cv_results_['std_test_score']
params=grid_result.cv_results_['params']

for mean,stdev,param in zip(means,stds,params):
    print('{0}  ({1}) with: {2}'.format(mean,stdev,param))

#The best result is with neuron1 and neuron2 8
```

```python
#predicting and calculating score for model
y_predict=grid.predict(x_standard)

from sklearn.metrics import classification_report, accuracy_score

print(accuracy_score(y,y_predict))
print(classification_report(y,y_predict))
```
